## Segemntig the data in to high, medium, low and very low vendors basing on the no of customers visiting


```python
categories = ['High', 'Medium', 'Low', 'Very Low']
categories_selected = [[] for i in range(len(categories))]
for name in data.columns:
    customer_visits = data[name].sum()
    if customer_visits > 100000:
        category = 0
    elif customer_visits > 40000:
        category = 1
    elif customer_visits > 10000:
        category = 2
    else:
        category = 3
    categories_selected[category].append(name)
    print('Vendor ' + name + ' is ' + categories[category] + ' Visits')
```

```python
for i, selected in enumerate(categories_selected):
    plt.figure(figsize=(8, 8))
    x_pos = np.arange(len(data[selected].columns))
    plt.bar(x_pos, data[selected].sum(), align='center')
    plt.xticks(x_pos, data[selected].columns)
    plt.xlabel('Vendors', fontsize=18)
    plt.ylabel('Customer_Visits', fontsize=18)
    plt.title(categories[i] + ' Total Customer Visits', fontsize=20)
    plt.show()
```

```python
colours = []
for name in data.columns:
    customer_visits = data[name].sum()
    if customer_visits > 100000:
        colour = 'green'
    elif customer_visits > 50000:
        colour = 'orange'
    elif customer_visits > 10000:
        colour = 'red'
    else:
        colour = 'black'
    colours.append(colour)

plt.figure(figsize=(17,8))
x_pos = np.arange(len(data.columns))
plt.bar(x_pos, data.sum(), align='center', color=colours)
plt.xticks(x_pos, data.columns)
plt.xlabel('Venues', fontsize=18)
plt.ylabel('Customer Visits', fontsize=18)
plt.title(' Total Volume of Customer Visits at Each Venue', fontsize=20)
plt.show()
```

```python
summary_data = pd.DataFrame(index=data.columns)
summary_data['age'] = visitors_age.values
summary_data['Distance'] = venue_Distance.values
summary_data['Duration'] = venue_Duration.values
summary_data['Avg_female'] = visitor_Gender.values
summary_data['Amount_spend'] = visitor_amountspend.values
summary_data['Customer_visits'] = data.sum().values

summary_data.head()
summary_data.describe()
```

```python
plt.figure(figsize=(10,10))
corr = summary_data.corr()
ax = sns.heatmap(corr, vmin=-1, vmax=1, center=0,
                 cmap=sns.diverging_palette(220,20,n=200),
                 square=True, annot=True, annot_kws={"size":10})
ax.set_xticklabels(ax.get_xticklabels(),rotation=45, horizontalalignment='right')
plt.xlabel('Summary Parameters',fontsize = 18)
plt.ylabel('Summary Parmeters', fontsize = 18)
plt.title('To Detect Correlation Between Summary Data', fontsize = 20)
plt.show()
```
